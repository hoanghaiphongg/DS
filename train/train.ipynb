```python
import pandas as pd
import statistics
from tqdm.notebook import tqdm

# Danh s√°ch m√πa gi·∫£i
seasons = ["2020-2021", "2021-2022", "2022-2023", "2023-2024", "2024-2025"]

# T·∫°o danh s√°ch c·∫ßu th·ªß cho ƒë·ªôi nh√† v√† ƒë·ªôi kh√°ch
home_players = ["home_starting_" + str(i) + "_name" for i in range(1, 12)]
away_players = ["away_starting_" + str(i) + "_name" for i in range(1, 12)]

# H√†m ph√¢n lo·∫°i c·∫ßu th·ªß theo v·ªã tr√≠
def get_position(team_player_ids):
    team = {"G": [], "M": [], "B": [], "F": []}  # Ch√∫ √Ω: B thay v√¨ D
    for player_id in team_player_ids:
        # B·ªè qua c·∫ßu th·ªß null ho·∫∑c NaN
        if pd.isna(player_id):
            continue
        
        # Chuy·ªÉn ID v·ªÅ d·∫°ng int n·∫øu l√† float
        if isinstance(player_id, float) and player_id.is_integer():
            player_id = int(player_id)
        
        # L·ªçc th√¥ng tin c·∫ßu th·ªß t·ª´ player_df
        player = players_df.loc[players_df["player_id"] == player_id]
        if player.empty:
            print(f"Kh√¥ng t√¨m th·∫•y c·∫ßu th·ªß ID: {player_id}")
            continue
            
        try:
            position = player["position"].values[0]
            if position in team:
                team[position].append(player_id)
            else:
                print(f"V·ªã tr√≠ kh√¥ng h·ª£p l·ªá: {position} cho c·∫ßu th·ªß {player_id}")
        except:
            print(f"L·ªói khi x·ª≠ l√Ω c·∫ßu th·ªß {player_id}")
            pass
    
    # Ki·ªÉm tra xem c√≥ ƒë·ªß c·∫ßu th·ªß ·ªü m·ªói v·ªã tr√≠ kh√¥ng
    check = True
    for pos in team:
        if not team[pos]:  # N·∫øu kh√¥ng c√≥ c·∫ßu th·ªß n√†o ·ªü v·ªã tr√≠ n√†y
            check = False
            print(f"Kh√¥ng c√≥ c·∫ßu th·ªß ·ªü v·ªã tr√≠ {pos}")
            break
    return team, check

# H√†m l·∫•y d·ªØ li·ªáu th·ªß m√¥n
def get_data_G(keeper_ids, season):
    if not keeper_ids:
        return [0] * len(gk_data)
        
    season_index = seasons.index(season)
    
    # Chuy·ªÉn ID v·ªÅ d·∫°ng int n·∫øu l√† float
    keeper_id = keeper_ids[0]
    if isinstance(keeper_id, float) and keeper_id.is_integer():
        keeper_id = int(keeper_id)
    
    # T√¨m d·ªØ li·ªáu t·ª´ m√πa hi·ªán t·∫°i, n·∫øu kh√¥ng c√≥ th√¨ t√¨m c√°c m√πa tr∆∞·ªõc
    while season_index >= 0:
        current_season = seasons[season_index]
        player = players_df.loc[(players_df["player_id"] == keeper_id) & 
                               (players_df["season"] == current_season)]
        
        if not player.empty:
            try:
                # Tr√≠ch xu·∫•t d·ªØ li·ªáu th·ªß m√¥n
                values = []
                for attr in gk_data:
                    if attr in player.columns:
                        values.append(player[attr].values[0])
                    else:
                        values.append(0)  # Gi√° tr·ªã m·∫∑c ƒë·ªãnh
                return values
            except Exception as e:
                print(f"L·ªói khi l·∫•y d·ªØ li·ªáu th·ªß m√¥n {keeper_id}: {e}")
                # Ti·∫øp t·ª•c v·ªõi m√πa tr∆∞·ªõc
        
        # Chuy·ªÉn sang m√πa tr∆∞·ªõc
        season_index -= 1
    
    # N·∫øu kh√¥ng t√¨m th·∫•y d·ªØ li·ªáu
    print(f"Kh√¥ng t√¨m th·∫•y d·ªØ li·ªáu cho th·ªß m√¥n {keeper_id}, s·ª≠ d·ª•ng gi√° tr·ªã m·∫∑c ƒë·ªãnh.")
    return [0] * len(gk_data)

# H√†m l·∫•y d·ªØ li·ªáu h·∫≠u v·ªá
def get_data_B(defender_ids, season):
    if not defender_ids:
        return [0] * len(back_data)
        
    season_index = seasons.index(season)
    values = []
    
    for player_id in defender_ids:
        # Chuy·ªÉn ID v·ªÅ d·∫°ng int n·∫øu l√† float
        if isinstance(player_id, float) and player_id.is_integer():
            player_id = int(player_id)
            
        try:
            current_season_index = season_index
            # T√¨m ki·∫øm d·ªØ li·ªáu t·ª´ m√πa hi·ªán t·∫°i ƒë·∫øn c√°c m√πa tr∆∞·ªõc
            while current_season_index >= 0:
                current_season = seasons[current_season_index]
                player = players_df.loc[(players_df["player_id"] == player_id) & 
                                       (players_df["season"] == current_season)]
                
                if not player.empty:
                    # Thu th·∫≠p d·ªØ li·ªáu c·∫ßu th·ªß
                    player_values = []
                    for attr in back_data:
                        if attr in player.columns:
                            player_values.append(player[attr].values[0])
                        else:
                            player_values.append(0)  # Gi√° tr·ªã m·∫∑c ƒë·ªãnh
                    values.append(player_values)
                    break
                
                # Chuy·ªÉn sang m√πa tr∆∞·ªõc
                current_season_index -= 1
            else:
                print(f"Kh√¥ng t√¨m th·∫•y d·ªØ li·ªáu cho h·∫≠u v·ªá {player_id}, s·ª≠ d·ª•ng gi√° tr·ªã m·∫∑c ƒë·ªãnh.")
                values.append([0] * len(back_data))
        except Exception as e:
            print(f"L·ªói khi x·ª≠ l√Ω h·∫≠u v·ªá {player_id}: {e}")
            values.append([0] * len(back_data))
    
    # N·∫øu kh√¥ng c√≥ gi√° tr·ªã n√†o
    if not values:
        return [0] * len(back_data)
        
    # T√≠nh trung b√¨nh c√°c ch·ªâ s·ªë c·ªßa t·∫•t c·∫£ h·∫≠u v·ªá
    aggregated_values = []
    for i in range(len(back_data)):
        aggregated_values.append(statistics.mean([val[i] for val in values]))
    
    return aggregated_values

# H√†m l·∫•y d·ªØ li·ªáu ti·ªÅn v·ªá (t∆∞∆°ng t·ª± nh∆∞ h·∫≠u v·ªá)
def get_data_M(midfielder_ids, season):
    if not midfielder_ids:
        return [0] * len(mid_data)
        
    season_index = seasons.index(season)
    values = []
    
    for player_id in midfielder_ids:
        # Chuy·ªÉn ID v·ªÅ d·∫°ng int n·∫øu l√† float
        if isinstance(player_id, float) and player_id.is_integer():
            player_id = int(player_id)
            
        try:
            current_season_index = season_index
            while current_season_index >= 0:
                current_season = seasons[current_season_index]
                player = players_df.loc[(players_df["player_id"] == player_id) & 
                                       (players_df["season"] == current_season)]
                
                if not player.empty:
                    # Thu th·∫≠p d·ªØ li·ªáu c·∫ßu th·ªß
                    player_values = []
                    for attr in mid_data:
                        if attr in player.columns:
                            player_values.append(player[attr].values[0])
                        else:
                            player_values.append(0)  # Gi√° tr·ªã m·∫∑c ƒë·ªãnh
                    values.append(player_values)
                    break
                
                current_season_index -= 1
            else:
                print(f"Kh√¥ng t√¨m th·∫•y d·ªØ li·ªáu cho ti·ªÅn v·ªá {player_id}, s·ª≠ d·ª•ng gi√° tr·ªã m·∫∑c ƒë·ªãnh.")
                values.append([0] * len(mid_data))
        except Exception as e:
            print(f"L·ªói khi x·ª≠ l√Ω ti·ªÅn v·ªá {player_id}: {e}")
            values.append([0] * len(mid_data))
    
    # N·∫øu kh√¥ng c√≥ gi√° tr·ªã n√†o
    if not values:
        return [0] * len(mid_data)
    
    # T√≠nh trung b√¨nh c√°c ch·ªâ s·ªë
    aggregated_values = []
    for i in range(len(mid_data)):
        aggregated_values.append(statistics.mean([val[i] for val in values]))
    
    return aggregated_values

# H√†m l·∫•y d·ªØ li·ªáu ti·ªÅn ƒë·∫°o
def get_data_F(forward_ids, season):
    if not forward_ids:
        return [0] * len(forward_data)
        
    season_index = seasons.index(season)
    values = []
    
    for player_id in forward_ids:
        # Chuy·ªÉn ID v·ªÅ d·∫°ng int n·∫øu l√† float
        if isinstance(player_id, float) and player_id.is_integer():
            player_id = int(player_id)
            
        try:
            current_season_index = season_index
            while current_season_index >= 0:
                current_season = seasons[current_season_index]
                player = players_df.loc[(players_df["player_id"] == player_id) & 
                                       (players_df["season"] == current_season)]
                
                if not player.empty:
                    # Thu th·∫≠p d·ªØ li·ªáu c·∫ßu th·ªß
                    player_values = []
                    for attr in forward_data:
                        if attr in player.columns:
                            player_values.append(player[attr].values[0])
                        else:
                            player_values.append(0)  # Gi√° tr·ªã m·∫∑c ƒë·ªãnh
                    values.append(player_values)
                    break
                
                current_season_index -= 1
            else:
                print(f"Kh√¥ng t√¨m th·∫•y d·ªØ li·ªáu cho ti·ªÅn ƒë·∫°o {player_id}, s·ª≠ d·ª•ng gi√° tr·ªã m·∫∑c ƒë·ªãnh.")
                values.append([0] * len(forward_data))
        except Exception as e:
            print(f"L·ªói khi x·ª≠ l√Ω ti·ªÅn ƒë·∫°o {player_id}: {e}")
            values.append([0] * len(forward_data))
    
    # N·∫øu kh√¥ng c√≥ gi√° tr·ªã n√†o
    if not values:
        return [0] * len(forward_data)
    
    # T√≠nh trung b√¨nh c√°c ch·ªâ s·ªë
    aggregated_values = []
    for i in range(len(forward_data)):
        aggregated_values.append(statistics.mean([val[i] for val in values]))
    
    return aggregated_values


# ƒê·ªãnh nghƒ©a c·∫•u tr√∫c d·ªØ li·ªáu ƒë·∫ßu ra
columns = ["match_id", "season"]

# Th√™m c·ªôt cho ƒë·ªôi nh√†
columns += [f"home_G_{attr}" for attr in gk_data]
columns += [f"home_B_{attr}" for attr in back_data]
columns += [f"home_M_{attr}" for attr in mid_data]
columns += [f"home_F_{attr}" for attr in forward_data]

# Th√™m c·ªôt cho ƒë·ªôi kh√°ch
columns += [f"away_G_{attr}" for attr in gk_data]
columns += [f"away_B_{attr}" for attr in back_data]
columns += [f"away_M_{attr}" for attr in mid_data]
columns += [f"away_F_{attr}" for attr in forward_data]

# Th√™m c·ªôt k·∫øt qu·∫£
columns.append("result")

# Kh·ªüi t·∫°o dataset v·ªõi header
data = [columns]

def create_training_data(limit=None):
    global data
    count = 0
    processed_count = 0
    
    # Ki·ªÉm tra ki·ªÉu d·ªØ li·ªáu tr∆∞·ªõc khi b·∫Øt ƒë·∫ßu
    print("Ki·ªÉm tra ki·ªÉu d·ªØ li·ªáu:")
    print("Ki·ªÉu d·ªØ li·ªáu c·ªßa c·ªôt player_id trong players_df:", players_df["player_id"].dtype)
    print("Ki·ªÉu d·ªØ li·ªáu trong matches_df:", matches_df[home_players[0]].dtype)
    
    # Duy·ªát qua t·ª´ng tr·∫≠n ƒë·∫•u
    for index, match in tqdm(matches_df.iterrows()):
        if limit is not None and count >= limit:
            break
            
        print(f"Processing match ID: {match['match_id']}")
        count += 1
        
        # Thu th·∫≠p ID c·∫ßu th·ªß
        home_team_ids = [match[player] for player in home_players]
        away_team_ids = [match[player] for player in away_players]
        
        # Ph√¢n lo·∫°i c·∫ßu th·ªß theo v·ªã tr√≠
        home_team_positions, home_check = get_position(home_team_ids)
        away_team_positions, away_check = get_position(away_team_ids)
        
        # B·ªè qua tr·∫≠n ƒë·∫•u n·∫øu kh√¥ng ƒë·ªß th√¥ng tin
        if not home_check or not away_check:
            print(f"B·ªè qua tr·∫≠n ƒë·∫•u {match['match_id']} do thi·∫øu th√¥ng tin c·∫ßu th·ªß")
            continue
        
        try:
            # L·∫•y d·ªØ li·ªáu cho ƒë·ªôi nh√†
            home_values = []
            home_values += get_data_G(home_team_positions["G"], match["season"])
            home_values += get_data_B(home_team_positions["B"], match["season"])
            home_values += get_data_M(home_team_positions["M"], match["season"])
            home_values += get_data_F(home_team_positions["F"], match["season"])
            
            # L·∫•y d·ªØ li·ªáu cho ƒë·ªôi kh√°ch
            away_values = []
            away_values += get_data_G(away_team_positions["G"], match["season"])
            away_values += get_data_B(away_team_positions["B"], match["season"])
            away_values += get_data_M(away_team_positions["M"], match["season"])
            away_values += get_data_F(away_team_positions["F"], match["season"])
            
            # X√°c ƒë·ªãnh k·∫øt qu·∫£ tr·∫≠n ƒë·∫•u
            if match["home_score"] > match["away_score"]:
                result = 0  # ƒê·ªôi nh√† th·∫Øng
            elif match["home_score"] == match["away_score"]:
                result = 1  # H√≤a
            else:
                result = 2  # ƒê·ªôi kh√°ch th·∫Øng
            
            # T·∫°o d√≤ng d·ªØ li·ªáu
            new_row = [match["match_id"], match["season"]] + home_values + away_values + [result]
            print("Successfully processed match")
            data.append(new_row)
            processed_count += 1
                
        except Exception as e:
            print(f"Error processing match {match['match_id']}: {e}")
    
    print(f"ƒê√£ x·ª≠ l√Ω th√†nh c√¥ng {processed_count}/{count} tr·∫≠n ƒë·∫•u")
    return data

# T·∫°o DataFrame t·ª´ d·ªØ li·ªáu ƒë√£ x·ª≠ l√Ω
def create_training_dataframe(data):
    # Chuy·ªÉn ƒë·ªïi danh s√°ch th√†nh DataFrame
    df = pd.DataFrame(data[1:], columns=data[0])
    return df


print("ƒêang x·ª≠ l√Ω to√†n b·ªô d·ªØ li·ªáu...")
data = create_training_data()  # B·ªè limit ƒë·ªÉ x·ª≠ l√Ω t·∫•t c·∫£ tr·∫≠n ƒë·∫•u
```
